1. Scrape the details of most viewed videos on YouTube from Wikipedia:

   Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
   
   You need to find following details:   
   A) Rank   
   B) Name   
   C) Artist   
   D) Upload date   
   E) Views

```python
driver = webdriver.Chrome(r"C:\Users\ABHINANDAN\Downloads\chromedriver_win32\chromedriver.exe")
#get the web page with given url
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)
#Lets fetch required data
rank = []
name = []
views = []
artist = []
upload_date = []
#lets fetch the rank
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")[:30]:
        rank.append(i.text)
except NoSuchElementException:
    rank.append('-')
#fetching video names
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]"):
        name.append(i.text)
except NoSuchElementException:
    name.append('-')
#fetching views
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")[:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append('-')
#fetching artist name
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")[:30]:
        artist.append(i.text)
except NoSuchElementException:        
    artist.append('-')
#fetching upload dates
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]"):
        upload_date.append(i.text)
except NoSuchElementException:        
    upload_date.append('-')
    
#lets create dataframe
data = list(zip(rank,name,views,artist,upload_date))
df = pd.DataFrame(data, columns = ["Rank", "Video Title", "Views (Billions)", "Artist Name", "Upload Date"])
df
```

5) Scrape the details of trending repositories on Github.com.
   Url = https://github.com/   
       You have to find the following details:   
       A) Repository title   
       B) Repository description   
       C) Contributors count   
       D) Language used
       
  Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
driver = webdriver.Chrome(r"C:\Users\ABHINANDAN\Downloads\chromedriver_win32\chromedriver.exe")
#get the web page with given url
url = "https://github.com/"
driver.get(url)
time.sleep(5)

#getting explore button
explore = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details").click()

#click on trending
trending = driver.find_element_by_xpath("/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a")

urls = trending.get_attribute("href")

driver.get(urls)

#getting url for every repository
rep_urls = []
repository = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']//a")
for i in repository:
    rep_urls.append(i.get_attribute("href"))

#Lets fetch repository title
rep_title = []
title = driver.find_elements_by_xpath("//h1[@class = 'h3 lh-condensed']")
for i in title:
    rep_title.append(i.text)

#fetch other data from every repository page
description = []
Contributors = []
language = []
for i in rep_urls:
    driver.get(i)
    time.sleep(3)
    #fetching decription
    try:
        desc = driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        description.append(desc.text)
    except NoSuchElementException:
        description.append('-')
    #fetching contribution count
    try:
        contri = driver.find_element_by_xpath("//*[contains(text(),'    Contributors ')]")
        Contributors.append(contri.text.replace('Contributors',''))
    except NoSuchElementException:
        Contributors.append('-')
    #fetching Languages used
    lang = []
    try:
        for i in driver.find_elements_by_xpath("//span[@class='color-text-primary text-bold mr-1']"):
            lang.append(i.text)
        language.append(lang)
    except NoSuchElementException:
        language.append('-')
        
#creating dataframe
data = list(zip(rep_title,description,Contributors,language))
df = pd.DataFrame(data, columns = [" Repository title","Repository description","Contributors count","Language used"])
df
```

10. __Details of Datasets from UCI machine learning repositories.__

    Url = https://archive.ics.uci.edu/
    
    You have to find the following details:    
    A) Dataset name    
    B) Data type    
    C) Task    
    D) Attribute type    
    E) No of instances    
    F) No of attribute    
    G) Year    
    Note: - from the home page you have to go to the Show All Dataset page through code.

```python
driver = webdriver.Chrome(r"C:\Users\ABHINANDAN\Downloads\chromedriver_win32\chromedriver.exe")
#get the web page with given url
url = " https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(3)

all_dataset = driver.find_element_by_xpath("//tbody[1]//tr/td[2]/span[2]/a")    #find view all dataset button 
page_url = all_dataset.get_attribute("href")               #getting url of page containing all datasets

driver.get(page_url)
time.sleep(5)

all_lst = driver.find_element_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[1]/tbody/tr/td[2]/p/a")   # view list button
lst_url = all_lst.get_attribute("href")           #getting page url containing list of all datasets
driver.get(lst_url)
time.sleep(5)

ds_url = driver.find_elements_by_xpath("//p[@class='normal']//b/a")    #fetching url for each dataset

urls = []                     #list of all urls for datasets
for i in ds_url:
    urls.append(i.get_attribute("href"))

#lets fetch the required data from every url
data_name = []
data_type = []
attr_type = []
Task = []
instances = []
n_attributes = []
Year = []
for i in urls:
    driver.get(i)
    time.sleep(3)
    #fetching Dataset name
    try: 
        ds_name = driver.find_element_by_xpath("//span[@class='heading']")
        data_name.append(ds_name.text)
    except NoSuchElementException:
        data_name.append('-')
    time.sleep(1)
    #fetching data type
    try:
        dtype = driver.find_element_by_xpath("//table[@border='1']//tbody/tr/td[2]")
        if dtype.text == "N/A": raise NoSuchElementException
        data_type.append(dtype.text)
    except NoSuchElementException:
        data_type.append('-')
    time.sleep(1)
    #fetching Attribute type
    try:
        atype = driver.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[2]")
        if atype.text == "N/A": raise NoSuchElementException
        attr_type.append(atype.text)
    except NoSuchElementException:
        attr_type.append('-')
    time.sleep(1)
    #fetching Task
    try:
        task = driver.find_element_by_xpath("//table[@border='1']//tbody/tr[3]/td[2]")
        if task.text == "N/A": raise NoSuchElementException
        Task.append(task.text)
    except NoSuchElementException:
        Task.append('-')
    time.sleep(1)
    # fetching No of instances
    try:
        inst = driver.find_element_by_xpath("//table[@border='1']//tbody/tr/td[4]")
        if inst.text == "N/A": raise NoSuchElementException
        instances.append(inst.text)
    except NoSuchElementException:
        instances.append('-')
    time.sleep(1)
    #fetching No of attribute
    try:
        attr = driver.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[4]")
        if attr.text == "N/A": raise NoSuchElementException
        n_attributes.append(attr.text)
    except NoSuchElementException:
        n_attributes.append('-')
    time.sleep(1)
    #fetching year
    try:
        year = driver.find_element_by_xpath("//table[@border='1']//tbody/tr[2]/td[6]")
        if year.text == "N/A": raise NoSuchElementException
        Year.append(year.text[:4])
    except NoSuchElementException:
        Year.append('-')
    time.sleep(1)
    
# lets create dataframe
data = list(zip(data_name,data_type,Task,attr_type,instances,n_attributes,Year))
df = pd.DataFrame(data, columns = ["Dataset Name","Data Type","Task","Attribute Type","Number of Instances","Number of Attributes","Year"])
df
```
